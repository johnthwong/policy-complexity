```python
incomes = list(range(0, 1_000_001, 1000))
mtr_list_by_increment = []

ecps_100["self_employment_income_last_year__2024"] = 0

sim = Microsimulation(
    dataset=ecps_100,
)

# for income in incomes:
#     ecps_100["employment_income_before_lsr__2024"] = income
#     sim = Microsimulation(
#         dataset=ecps_100,
#     )
#     mtr = sim.calculate("marginal_tax_rate", period=2026).to_frame()
#     mtr_list_by_increment.append(mtr)

for income in incomes:
    sim.set_input("employment_income", 2026, [income]*254)
    inc = sim.calculate("employment_income", period=2026).to_frame()
    mtr = sim.calculate("marginal_tax_rate", period=2026).to_frame()
    mtr_list_by_increment.append(mtr)
    mtr_list_by_increment.append(inc)
```
