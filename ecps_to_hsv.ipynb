```python
import pandas as pd
```

```python
from policyengine_us import Microsimulation
ECPS = "hf://policyengine/policyengine-us-data/enhanced_cps_2024.h5"
sim = Microsimulation(dataset=ECPS)
sim.subsample(100, seed = 1000) # .subsample is an in-place method
df = sim.to_input_dataframe()
```

```python
df
```

```python
weights = sim.calculate("household_weight", period=2024).to_list()
weights
```

```python
num = pow(sum(weights), 2)
denom = 0

for weight in weights:
    denom+=pow(weight, 2)

num/denom
```

```python
sim2 = Microsimulation(
    dataset=df,
)
dependence = sim2.calculate("is_tax_unit_dependent", period = 2024).to_frame()
df2 = pd.concat([df, dependence], axis = 1)
```

```python
df2.rename(
    columns={
        'spm_unit_net_income_reported__2024': 'net_income', 
        'spm_unit_total_income_reported__2024': 'income',
        0: 'is_tax_unit_dependent'
        }, 
    inplace=True
    )
```

```python
df2.columns
```

```python
spm_demographics = []
for spm_unit in list(set(df2["spm_unit_id__2024"])):
    members_list = df2[df2["spm_unit_id__2024"]==spm_unit]
    dependents = sum(members_list["is_tax_unit_dependent"])
    household_size = members_list.shape[0]
    working = household_size - dependents
    spm_demographics.append([spm_unit, household_size, working, dependents])

```

```python
df_spm_demographics = pd.DataFrame(
    spm_demographics,
    columns=['spm_unit', 'household_size', 'working', 'dependents']
    )
```

```python
df3 = pd.merge(
    df2, df_spm_demographics, 
    how='left', 
    left_on='spm_unit_id__2024',
    right_on='spm_unit'
    )

df4 = df3[
    (df3["is_household_head__ETERNITY"] == True) &
    (df3["income"] > 0) &
    (df3["net_income"] > 0)
]
```
